```python
import numpy as np
import pandas as pd
import seaborn as sns
dataset=pd.read_csv('C:/Users/Pranab Mahata/Desktop/ML/data (3).csv')
dataset.head(3)
```

In this project our task is to predict whether anyone is affacetd by Chronic_kidney_disease and	Adrenal_and_thyroid disorders or not.

```python
# Remove null values and clean preprocess the data
dataset=dataset.dropna(axis=0)
# axis =0 means the null values will be removed by row-wise .
# axis =1 means the null values will be removed by column-wise.
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.corr()
```

If the correlation score between the features is positive then they are in the same direction. But if it is negative then it is in opposite direction.

```python
sns.heatmap(dataset.corr(), annot=True)
```

```python
x=dataset.iloc[:,:-2]
x.head(3)
# feature data
```

```python
y1=dataset.iloc[:,-1]
print(y1.head(3))

y2=dataset.iloc[:,-2]
print(y2.head(3))

#target column
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y1,test_size=0.2)
m_train,m_test,n_train,n_test=train_test_split(x,y2,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn=KNeighborsClassifier(n_neighbors=7)
knn_model_1=knn.fit(x_train,y_train)
knn_model_2=knn.fit(m_train,n_train)
knn_predict_1=knn.predict(x_test)
knn_predict_2=knn.predict(m_test)
print("First disease: Adrenal_and_thyroid_disorders")
print()
print("Second disease: Chronic_kidney_disease ")
print()
print(knn_predict_1)
print()
print(knn_predict_2)
print()
print("Accuracy score for the possibility of two different diseases: ")
print()
print(metrics.accuracy_score(y_test,knn_predict_1))
print(metrics.accuracy_score(n_test,knn_predict_2))
print()
print("Overall accuracy: ",(metrics.accuracy_score(y_test,knn_predict_1)+metrics.accuracy_score(n_test,knn_predict_2))/2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy', max_depth=10)
dtree_model_1=dtree.fit(x_train,y_train)
dtree_model_2=dtree.fit(m_train,n_train)
dtree_predict_1=dtree.predict(x_test)
dtree_predict_2=dtree.predict(m_test)
print("First disease: Adrenal_and_thyroid_disorders")
print()
print("Second disease: Chronic_kidney_disease ")
print()
print(dtree_predict_1)
print()
print(dtree_predict_2)
print()
print("Accuracy score for the possibility of two different diseases: ")
print()
print(metrics.accuracy_score(y_test,dtree_predict_1))
print(metrics.accuracy_score(n_test,dtree_predict_2))
print()
print("Overall accuracy: ", (metrics.accuracy_score(y_test,dtree_predict_1)+metrics.accuracy_score(n_test,dtree_predict_2))/2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=50,criterion='entropy', max_depth=10)
rf_model_1=rf.fit(x_train,y_train)
rf_model_2=rf.fit(m_train,n_train)
rf_predict_1=rf.predict(x_test)
rf_predict_2=rf.predict(m_test)
print("First disease: Adrenal_and_thyroid_disorders")
print()
print("Second disease: Chronic_kidney_disease ")
print()
print(rf_predict_1)
print()
print(rf_predict_2)
print()
print("Accuracy score for the possibility of two different diseases: ")
print()
print(metrics.accuracy_score(y_test,rf_predict_1))
print(metrics.accuracy_score(n_test,rf_predict_2))
print()
print("Overall accuracy: ", (metrics.accuracy_score(y_test,rf_predict_1)+metrics.accuracy_score(n_test,rf_predict_2))/2)
```

```python
knn_model_1_cm=metrics.confusion_matrix(y_test,knn_predict_1)
knn_model_2_cm=metrics.confusion_matrix(n_test,knn_predict_2)
print("First disease: Adrenal_and_thyroid_disorders")
print()
print("Second disease: Chronic_kidney_disease ")
print()
print(knn_model_1_cm)
print()
print(knn_model_2_cm)
print()
print("Class for Adernal_and_thyroid_disorders is: 1 and 0")
print()
print("Class for Chronic_kidney_disease is: 1 and 0")
```
